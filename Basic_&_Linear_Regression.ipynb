### Regression
- 회귀 분석은 데이터 값이 평균과 같은 일정한 값으로 돌아기려는 경향을 이용한 통계학 기법이다
- 머신러닝 회귀 예측의 핵심은 주어진 피처와 결정 값 데이터 기반에서 학습을 통해 최적의 회귀 계수 (parameter)를 찾아내는 것이다 

### Linear Regression
- 실제 값과 예측값의 차이를 최소화하는 직선형 회귀선을 최적화하는 방식이다


#### 가설 함수 (Hypothesis function)
- 데이터의 특징에 가장 잘 맞는 최적의 선을 찾으려고 시도해보는 모든 함수를 의미 
- 학습: 데이터에 가장 잘 맞는 함수를 찾는 것
- y = ax + b
- Model Capacity (역량)
  - 모델의 데이터에 대한 설명력
  - parameter 개수 up -> 설명력 up
- First-order model (1차함수)
  - feature가 하나인 경우: hθ(x) = θ0 + θ1x1
  - feature가 여러 개인 경우: hθ(x) = θ0 + θ1x1 + θ2x2 + … + θnxn
    - 다중회귀: model capacity up!
- High-order model: 2차함수 이상 (다항회귀)

- 다항 & 다중
  - 다항이 곡선을 만들고
  - 다중이 면을 만들어
  - 다항과 다중 회귀가 합쳐지면 면이 구부러진다

#### 손실함수: MSE (평균제곱오차)
- Loss Function: 가설 함수를 평가하기 위한 함수. 손실 함수의 결과값이 작을수록 손실이 작은 것 = 좋은 가설 함수
- 손실함수의 최저점, 즉 MSE가 가장 최소가 되는 지점을 찾으면 데이터에 가장 잘 맞는 가설함수를 찾을 수 있다.
- MSE: 각 데이터포인트들이 가설 함수로부터 얼마나 떨어져 있는지, 그 오차를 모두 제곱해서 합하고 평균을 낸것
  - MSE: 실제값(y)과 예측값(ŷ)의 차이를 비교
  - mean(y- ŷ)^2 = mean(y-wx+b)^2

#### Gradient Descent (경사하강법)
- 특정 지점에서 시작하여 MSE의 최저점을 찾는 방향으로 step size를 갖고 parameter를 자동을 update 시키는 방식
- 손실함수의 최저점을 찾는 함수
- process
  - 무작위 시작점을 정하고 가설함수의 w와 b의 값 (parameter)을 조금씩 바꾸면서 손실함수 (MSE)를 줄인다 
  - 매번 parameter를 바꿀 때마다 MSE를 가장 크게 줄인 경사도를 택하면 되고
  - 경사도(기울기)는 편미분을 통해 구한다

### ii. Modeling

#### 1) Integer Encoding
- object to int64
- sex, smoker, region to int64

```python
from sklearn.preprocessing import LabelEncoder

encoder1 = LabelEncoder()
DF['sex'] = encoder1.fit_transform(DF.sex)
encoder2 = LabelEncoder()
DF['smoker'] = encoder2.fit_transform(DF.smoker)
encoder3 = LabelEncoder()
DF['region'] = encoder3.fit_transform(DF.region)
```

```python
DF.head(3)
```

#### 2) Train & Test Split (7:3) (다중회귀)

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split

X = DF[['age', 'sex']] ## 다중회귀
y = DF['expenses']

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.3,
                                                    random_state = 2045)

print('Train Data : ', X_train.shape, y_train.shape)
print('Test Data : ', X_test.shape, y_test.shape)
```

#### 3)  N차 다항식 변환 (다항회귀)

```python
# from sklearn.preprocessing import PolynomialFeatures

# poly = PolynomialFeatures(degree = 2, include_bias = False) # N =2
# X_train_poly = poly.fit_transform(X_train)

# print('변환 전 데이터: ', X_train.shape)
# print('2차항 변환 데이터: ', X_train_poly.shape)
```

#### 4) .fit(): learning (경사하강)

```python
from sklearn.linear_model import LinearRegression

NL = LinearRegression()
NL.fit(X_train, y_train)
```

#### 5) w & b (parameter) 값 확인

```python
import numpy as np
np.set_printoptions(suppress = True, precision = 10)

print('weight(w) : ', NL.coef_)
print('bias(b) : ', '%.8f' % NL.intercept_)
```

- 가설함수가 다중 & 다항에 따라서 w를 5개 가지게 됨 

#### 6) .predict(): ŷ 값

```python
y_hat = NL.predict(X_test)
```

#### 7) MSE

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mse2 = mean_squared_error(y_test, y_hat)

np.sqrt(mse2) # 의료비에 가깝게 만들기에 (너무 커서)
```

#### 8) R-Square (결정계수)

```python
# X_test_poly = poly.fit_transform(X_test)

NL.score(X_test, y_test)
```

> ## 1) sigmoid( ) 정의

```
import numpy as np

def sigmoid(x):
    y_hat = 1 / (1 + np.exp(-x))
    return y_hat
```

> ## 2) sigmoid( ) 실행

```
sigmoid(0)
```

```
sigmoid(100000000)
```

```
sigmoid(-100000000)
```

> ## 3) sigmoid( ) 시각화

```
import matplotlib.pyplot as plt

n = np.linspace(-10.0, 10.0, 2000)

plt.figure(figsize = (9, 6))
plt.plot(n, sigmoid(n))
plt.show()
```
